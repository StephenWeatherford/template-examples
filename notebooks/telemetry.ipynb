# Survey

```polyglot-notebook
RawEventsVSCodeExt
| where ServerTimestamp >= datetime(2023-02-13T15:03:21.06Z)
| where ExtensionName == "vscode-bicep"
| where EventName == "vscode-bicep/survey"
| summarize Count=count() by SurveyLinkStatus=tostring(Properties.surveylinkstatus)
| order by Count
```

# Linter rules

```polyglot-notebook

// Changing rule levels
// Rule's level changed via editor, by rule and to which level
RawEventsVSCodeExt
| where ExtensionName == "vscode-bicep"
| where EventName == "vscode-bicep/linter/rulestatechange"
| extend Rule = tostring(Properties["rule"]), FromLevel=tostring(Properties["previousdiagnosticlevel"]), ToLevel=tostring(Properties["currentdiagnosticlevel"])
| summarize HowOftenChangedInsideEditor=count () by Rule, ToLevel
| order by HowOftenChangedInsideEditor desc

// Rule's level changed via editor, by rule, from level and to level
RawEventsVSCodeExt
| where ExtensionName == "vscode-bicep"
| where EventName == "vscode-bicep/linter/rulestatechange"
| extend Rule = tostring(Properties["rule"]), FromLevel=tostring(Properties["previousdiagnosticlevel"]), ToLevel=tostring(Properties["currentdiagnosticlevel"])
| summarize HowOftenChangedInsideEditor=count () by Rule, FromLevel, ToLevel
| order by HowOftenChangedInsideEditor desc

// Rules most frequently set to "off"
RawEventsVSCodeExt
| where ExtensionName == "vscode-bicep"
| where EventName == "vscode-bicep/linter/rulestatechange"
| extend Rule = tostring(Properties["rule"]), FromLevel=tostring(Properties["previousdiagnosticlevel"]), ToLevel=tostring(Properties["currentdiagnosticlevel"])
| where ToLevel == "off"
| summarize HowOftenTurnedOffInsideEditor=count () by Rule
| order by HowOftenTurnedOffInsideEditor desc
```
